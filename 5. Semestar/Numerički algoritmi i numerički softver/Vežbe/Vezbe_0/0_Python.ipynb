## Osnovni tipovi podataka

U Pythonu najčešće koristimo:
- **int** – celi brojevi (npr. `-2`, `0`, `42`)
- **float** – realni brojevi sa decimalama (npr. `3.14`, `-0.5`)
- **bool** – logičke vrednosti (`True`, `False`)
- **str** – tekstualni niz (string), npr. `"Hello"` ili `'Ćao'`
- **None** – specijalna vrednost koja označava „nema vrednosti“

Korisne funkcije i koncepti:
- `type(x)` vraća tip promenljive `x`
- **konverzije**: `int("42")`, `float("3.14")`, `str(123)`, `bool(x)`
- **f-string** za formatiranje: `f"Rezultat je {vrednost}"`

Napomene:
- Deljenje `/` uvek vraća `float`; celobrojno deljenje `//` vraća `int`
- Operator `%` je ostatak pri deljenju
- Poređenja vraćaju `bool` (`==`, `!=`, `<`, `<=`, `>`, `>=`)
- "Istinitost" (<i>truthiness</i>): prazni nizovi/0/`None` se tretiraju kao `False`


````python
# Osnovni tipovi i osnovne operacije

# int, float, bool, str, None
a = 42              # int
b = 3.5             # float
c = True            # bool
d = "Python"        # str
e = None            # None

print("TIPOVI:")
for name, val in [("a", a), ("b", b), ("c", c), ("d", d), ("e", e)]:
    print(f"{name} = {val!r:>8}  -> type: {type(val).__name__}")

# Aritmetika
x = 7
y = 3
print("\nARITMETIKA:")
print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)     # uvek float
print("x // y =", x // y)   # celobrojno deljenje
print("x % y =", x % y)     # ostatak
print("x ** y =", x ** y)   # stepenovanje

# Zaokruživanje
print("\nZAOKRUŽIVANJE:")
print("round(3.14159, 2) =", round(3.14159, 2))

# Poređenja -> bool
print("\nPOREĐENJA:")
print("5 == 5 :", 5 == 5)
print("5 != 3 :", 5 != 3)
print("2 < 10 :", 2 < 10)
print("10 <= 10:", 10 <= 10)

# Konverzije tipova
print("\nKONVERZIJE:")
print('int("42") ->', int("42"))
print('float("3.14") ->', float("3.14"))
print("str(123) ->", str(123))
print("bool(0) ->", bool(0))
print("bool('') ->", bool(''))
print("bool('tekst') ->", bool("tekst"))

# Rad sa stringovima
ime = "Ana"
godine = 25
print("\nSTRINGOVI:")
print("Konkatenacija:", "Zdravo, " + ime + "!")
print(f"f-string: {ime} ima {godine} godina.")
print("Dužina stringa 'Python':", len(d))
print("Mala slova:", d.lower(), "| Velika slova:", d.upper())
print("Substring 'Py' u 'Python'?", "Py" in d)

# Provera tipa
print("\nPROVERA TIPA:")
print("isinstance(42, int) ->", isinstance(42, int))
print("isinstance(3.5, (int, float)) ->", isinstance(3.5, (int, float)))

# Truthiness primeri
print("\nTRUTHINESS:")
for val in [0, 1, "", " ", [], [1], None]:
    print(f"{val!r:>6} -> {bool(val)}")
````

## Kontrola toka – if / elif / else / while

U Pythonu se **grananje** i **ponavljanje** rade pomoću `if`, `elif`, `else` i `while`.

---
### if / elif / else

Služi za donošenje odluka na osnovu uslova.  
Sintaksa:

```python
if uslov1:
    # blok koda ako je uslov1 tačan
elif uslov2:
    # ako prvi nije, ali ovaj jeste
else:
    # ako nijedan uslov nije tačan
```
---
#### Kombinovanje uslova

Uslove možemo kombinovati pomoću logičkih operatora:
- and – oba uslova moraju biti tačna
- or – bar jedan uslov mora biti tačan
- not – negira logičku vrednost
---
#### Truthiness (istinitost vrednosti)

U Pythonu se određene vrednosti automatski smatraju „neistinama“ (False) u uslovima.
To uključuje:
- 0, 0.0
- "" (prazan string)
- [], {}, set() (prazne kolekcije)
- None

Sve ostale vrednosti tretiraju se kao True

---
### WHILE petlja

while petlja izvršava blok koda dokle god je uslov tačan (True).
```python
while uslov:
    # ponavljaj dok je uslov True
```
---
#### break i continue

- break – prekida izvršavanje petlje
- continue – preskače ostatak trenutne iteracije i prelazi na sledeću



````python
# Osnovni if/elif/else tok
x = 7

if x > 10:
    print("x je veće od 10")
elif x == 10:
    print("x je tačno 10")
else:
    print("x je manje od 10")

# Kombinovanje uslova
a = 5
b = 8
if a < b and b < 10:
    print("a je manje od b, a b je manje od 10")

if not (a > 10 or b < 5):
    print("Uslov nije ispunjen, pa se izvršava ovaj deo.")

# Truthiness
if []:
    print("Lista nije prazna")
else:
    print("Lista je prazna")

# Ternarni (skraćeni) if izraz
broj = -3
rezultat = "pozitivan" if broj > 0 else "negativan"
print("Broj je:", rezultat)
````

````python
# While petlja
brojac = 1
while brojac <= 5:
    print("Brojač:", brojac)
    brojac += 1  # povećava se svaki put

# break i continue
n = 0
while n < 10:
    n += 1
    if n == 3:
        continue  # preskače ostatak i ide dalje
    if n == 7:
        break     # prekida petlju
    print("n =", n)

print("Kraj petlje.")
````

## Liste (Lists)

**Liste** su najčešće korišćena kolekcija u Pythonu.  
Predstavljaju **uređenu, promenljivu sekvencu elemenata**, koja može da sadrži **različite tipove podataka**.

Osnovne osobine:
- Definišu se uglastim zagradama: `[]`
- Elementi se razdvajaju zarezima
- Indeksi počinju od **0**
- Mogu se menjati (mutable)
- Mogu biti ugnježdene (liste u listi)

Korisne funkcije i operacije:
- `len(lista)` – broj elemenata
- Dodavanje: `append(x)`, `extend([a,b])`, `insert(i, x)`
- Uklanjanje: `remove(x)`, `pop(i)`, `clear()`
- Sortiranje: `sort()`, `reverse()`
- Slicing: `lista[start:end:korak]`
- Provera: `x in lista`


````python
# Liste – primeri osnovnih operacija

# Kreiranje liste
brojevi = [10, 20, 30, 40]
mesovita = [1, "Python", 3.14, True]

print("LISTE:")
print("brojevi =", brojevi)
print("mesovita =", mesovita)
print("Broj elemenata:", len(brojevi))

# Pristup elementima
print("\nPRISTUP ELEMENTIMA:")
print("Prvi element:", brojevi[0])
print("Poslednji element:", brojevi[-1])

# Promena vrednosti
brojevi[1] = 25
print("Izmenjena lista:", brojevi)

# Dodavanje elemenata
brojevi.append(50)
print("Nakon append:", brojevi)
brojevi.insert(1, 15)
print("Nakon insert:", brojevi)

# Uklanjanje elemenata
brojevi.remove(25)
print("Nakon remove(25):", brojevi)
izbacen = brojevi.pop()
print("Nakon pop():", brojevi, "| Izbačen:", izbacen)

# Slicing
print("\nSLICE:")
print("Prva tri elementa:", brojevi[:3])
print("Svaki drugi element:", brojevi[::2])

# Konkatenacija i ponavljanje
print("\nKONKATENACIJA I PONAVLJANJE:")
print("brojevi + [60, 70] ->", brojevi + [60, 70])
print("brojevi * 2 ->", brojevi * 2)

# Sortiranje i obrnut redosled
lista = [3, 1, 4, 1, 5, 9]
print("\nSORTIRANJE:")
lista.sort()
print("Sortirano:", lista)
lista.reverse()
print("Obrnuto:", lista)

# Ugnježdene liste
matrica = [[1, 2], [3, 4], [5, 6]]
print("\nUGNJEŽDENE LISTE:")
print("matrica =", matrica)
print("Pristup elementu [1][0]:", matrica[1][0])

# Iteracija
print("\nITERACIJA:")
for x in brojevi:
    print(x, end=" ")

# List comprehension
print("\n\nLIST COMPREHENSION:")
kvadrati = [x**2 for x in range(1, 6)]
print("Kvadrati:", kvadrati)
````

## Rečnici (Dictionaries)

**Rečnik (`dict`)** je kolekcija podataka koja čuva **parove ključeva i vrednosti**.  
Ključevi moraju biti **jedinstveni i nepromenljivi (immutable)** tipovi podataka — najčešće stringovi ili brojevi.  
Vrednosti mogu biti bilo koji tip (brojevi, stringovi, liste, drugi rečnici itd).

---

### Osnovne osobine

- Definiše se pomoću vitičastih zagrada `{}`  
- Svaki element ima oblik `ključ: vrednost`  
- Ključevi su jedinstveni  
- Redosled unosa se pamti (od Python 3.7 nadalje)  
- Može se menjati (mutable)


````python

# Kreiranje rečnika

# Prazan rečnik
prazan = {}

# Rečnik sa vrednostima
osoba = {
    "ime": "Ana",
    "godine": 25,
    "grad": "Beograd"
}

print(osoba)

# Pristup pomoću ključa
print(osoba["ime"])        # Ana

# Ako ključ ne postoji -> KeyError
# Bez greške možemo koristiti get()
print(osoba.get("zanimanje"))       # None
print(osoba.get("zanimanje", "Nepoznato"))  # Nepoznato

osoba["zanimanje"] = "programer"    # dodavanje novog ključa
osoba["godine"] = 26                # izmena postojećeg ključa
print(osoba)

osoba.pop("grad")   # uklanja ključ 'grad'
print(osoba)

del osoba["zanimanje"]   # drugačiji način uklanjanja
print(osoba)

osoba.clear()   # briše sve elemente
print(osoba)

student = {
    "ime": "Milan",
    "prosek": 9.2,
    "smer": "Računarstvo"
}

# Iteracija kroz ključeve
for kljuc in student:
    print(kljuc)

# Iteracija kroz vrednosti
for vrednost in student.values():
    print(vrednost)

# Iteracija kroz parove (ključ, vrednost)
for kljuc, vrednost in student.items():
    print(f"{kljuc} -> {vrednost}")

a = {"x": 1, "y": 2}
b = {"y": 3, "z": 4}

# Metoda update() spaja i ažurira vrednosti
a.update(b)
print(a)  # {'x': 1, 'y': 3, 'z': 4}

# Od Pythona 3.9 moguće i pomoću operatora |
c = a | {"w": 5}
print(c)

studenti = {
    "1": {"ime": "Ana", "poeni": 85},
    "2": {"ime": "Jovan", "poeni": 92},
    "3": {"ime": "Ivana", "poeni": 78}
}

print(studenti["2"]["ime"])  # Jovan

# Iteracija kroz ugnježdene rečnike
for id, podaci in studenti.items():
    print(f"{id}: {podaci['ime']} ({podaci['poeni']} poena)")
````

## Funkcije u Pythonu

**Funkcije** su blokovi koda koji se mogu višestruko koristiti.  
Koriste se za grupisanje logike, izbegavanje ponavljanja i čine program organizovanijim i čitljivijim.

---

### Osnovna definicija

Funkcija se definiše pomoću ključne reči `def`, a može (ali ne mora) imati parametre i `return` vrednost.

```python
def pozdrav():
    print("Zdravo iz funkcije!")

pozdrav()
```

````python
def saberi(a, b):
    return a + b

print(saberi(5, 3))

# Podrazumevani i imenovani argumenti
def pozdrav(ime="Gost"):
    print(f"Zdravo, {ime}!")

pozdrav("Ana")
pozdrav()

def predstavi_osobu(ime, godine, grad):
    print(f"{ime} ima {godine} godina i živi u {grad}.")

predstavi_osobu(grad="Beograd", ime="Milan", godine=25)

# Promenljiv broj argumenata
def prikazi(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

prikazi(1, 2, 3, ime="Ana", poeni=95)

# Lambda funkcije
kvadrat = lambda x: x**2
print(kvadrat(5))

# Više povratnih vrednosti
def racunaj(a, b):
    return a+b, a-b, a*b

s, o, m = racunaj(10, 3)
print(s, o, m)
````

## Klase i objekti u Pythonu

Python podržava **objektno orijentisano programiranje (OOP)**.  
Klase služe kao šabloni za kreiranje **objekata**, koji imaju **podatke (atribute)** i **ponašanja (metode)**.

---

### Osnovna definicija

```python
class Osoba:
    def __init__(self, ime, godine):
        self.ime = ime
        self.godine = godine

    def predstavi_se(self):
        print(f"Zdravo, ja sam {self.ime} i imam {self.godine} godina.")

# Kreiranje objekta
o = Osoba("Ana", 25)
o.predstavi_se()
```
---
### Atributi i metode

- Atributi: promenljive koje pripadaju objektu (self.ime)

- Metode: funkcije definisane unutar klase

- self označava trenutni objekat i obavezno je u metodama instance
---
### Nasleđivanje

- Klasa može naslediti osobine i metode druge klase.

````python
# Osnovna definicija
class Osoba:
    def __init__(self, ime, godine):
        self.ime = ime
        self.godine = godine

    def predstavi_se(self):
        print(f"Zdravo, ja sam {self.ime} i imam {self.godine} godina.")

# Kreiranje objekta
o = Osoba("Ana", 25)
o.predstavi_se()

# Atributi i metode
class Krug:
    def __init__(self, r):
        self.r = r

    def povrsina(self):
        return 3.14 * self.r ** 2

k = Krug(5)
print("Površina:", k.povrsina())

# Nasleđivanje
class Zaposleni(Osoba):
    def __init__(self, ime, godine, pozicija):
        super().__init__(ime, godine)
        self.pozicija = pozicija

    def predstavi_se(self):
        print(f"{self.ime}, {self.pozicija}, {self.godine} godina")

z = Zaposleni("Jovan", 30, "Programer")
z.predstavi_se()
````

## NumPy i Matplotlib biblioteke

NumPy i Matplotlib su osnovne biblioteke za naučno računanje i vizuelizaciju u Pythonu.

---

### NumPy – rad sa nizovima (arrays)

**NumPy** omogućava efikasno rukovanje velikim numeričkim podacima kroz višedimenzionalne nizove.


````python
### Kreiranje i osnovne operacije
import numpy as np

a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])

print("a =", a)
print("Dimenzije:", a.ndim)
print("Oblik:", b.shape)
````

````python
# Tipovi podataka i indeksiranje
print("Tip elemenata:", a.dtype)
print("Prvi element:", a[0])
print("Podniz:", b[:, 1])   # druga kolona
````

````python
# Matematicke operacije
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("Zbir:", x + y)
print("Skalarni proizvod:", x * y)
print("Kvadrat:", x ** 2)
print("Srednja vrednost:", np.mean(x))
````

````python
# Rad sa matricama
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("A:")
print(A)
print("B:")
print(B)

# Matrični proizvod
C = np.dot(A, B)
print("A · B =\n", C)

# Alternativno korišćenjem @ operatora
print("A @ B =\n", A @ B)

# Transponovanje i determinanta
print("Transponovana A:\n", A.T)
print("Determinanta A:", np.linalg.det(A))
````

````python
# Višedimenzionalni nizovi
mat = np.arange(1, 10).reshape(3, 3)
print(mat)
print("Transponovana matrica:\n", mat.T)
````

## Matplotlib – vizuelizacija podataka

Matplotlib služi za crtanje dijagrama i grafika u Pythonu.

````python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sinusna funkcija")
plt.xlabel("x osa")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
````

````python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x), 'r')
plt.title("sin(x)")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), 'b')
plt.title("cos(x)")

plt.tight_layout()
plt.show()
````
