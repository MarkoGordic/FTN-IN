# Primer p-vrednosti kroz bacanje novčića: dodatno objašnjenje

Zamislimo da želimo da proverimo da li je novčić pošten, odnosno da li daje Head i Tail sa jednakom verovatnoćom od 50%.  
Da bismo to uradili, bacamo novčić određeni broj puta i posmatramo rezultate.

U statistici postavljamo hipoteze:

- **H₀ (nul-hipoteza):** novčić je pošten (Head = 50%)
- **H₁ (alternativna hipoteza):** novčić nije pošten

**Nul-hipoteza** predstavlja osnovnu pretpostavku koju statistički test *pokušava da obori*.  
To je polazna tačka; pretpostavljamo da je novčić pošten, a p-vrednost nam govori da li posmatrani podaci podržavaju tu pretpostavku ili je dovode u pitanje.

Kada završimo eksperiment i izbrojimo koliko puta smo dobili Head, koristimo p-vrednost da procenimo da li se naš rezultat uklapa u ono što bismo očekivali od poštenog novčića.

---

## Šta p-vrednost zapravo znači?

P-vrednost predstavlja verovatnoću da bismo dobili rezultat **bar toliko ekstreman kao onaj koji smo posmatrali**, i to **pod pretpostavkom da je novčić zaista pošten**.

Drugim rečima:

**P-vrednost meri koliko je naš rezultat “čudan” ili “neuobičajen” ako je novčić pošten.**

Ako je p-vrednost mala, to znači da bi ovakav rezultat bio veoma malo verovatan za pošten novčić.

---

## Primer

Pretpostavimo da smo bacili novčić 20 puta i dobili 17 Head.  
Od poštenog novčića očekivali bismo oko 10 Head, možda malo više ili manje, ali 17 deluje kao jako veliko odstupanje.

P-vrednost nam tada govori:

**Ako je novčić zaista pošten, kolika je verovatnoća da se desi ovako ekstreman rezultat?**

**p ≈ 0.0026** (neću pisati ovde kako se računa, ovo bi zaista trebalo da obradite na statistici sledeći semestar :))

Ovo znači da bi, ako je novčić zaista pošten, šansa da posmatramo rezultat ovako ekstreman ili ekstremniji bila oko 0.26%.

---

## Kako se donosi zaključak?

U praksi se često koristi prag od **0.05**:

- Ako je **p ≥ 0.05**, rezultat se smatra uobičajenim → zadržava se hipoteza da je novčić pošten.
- Ako je **p < 0.05**, rezultat je previše ekstreman → odbacuje se ideja da je novčić pošten.

U našem primeru p-vrednost je mnogo manja od 0.05, pa zaključujemo:

**Posmatrani rezultat je toliko neuobičajen da verovatno ne potiče od poštenog novčića.**

---

## Suština

P-vrednost nam ne govori da li je novčić *sigurno* nepošten.  
Umesto toga, govori nam koliko su naši podaci kompatibilni sa pretpostavkom da je novčić pošten.

- Velika p-vrednost → podaci se dobro uklapaju u tu pretpostavku.  
- Mala p-vrednost → podaci se ne uklapaju i sugerišu da pretpostavka verovatno nije tačna.

Drugim rečima:

**P-vrednost je alat koji nam pokazuje da li su rezultati eksperimenta “normalni” ili “previše čudni” u odnosu na ono što očekujemo.**


# Tumačenje p-vrednosti u statističkim testovima nad rezidualima

Pri analizi regresionih modela proveravamo različite pretpostavke o rezidualima, kao što su normalnost i jednaka varijansa.  
Sve ove provere funkcionišu po istom principu kao i primer sa bacanjem novčića: postavimo **nul-hipotezu**, izračunamo **p-vrednost**, i na osnovu nje procenjujemo da li posmatrani podaci podržavaju početnu pretpostavku.

U oba testa koristi se prag od 0.05.  
Ako je p-vrednost manja od tog praga, rezultat smatramo dovoljno “neuobičajenim” da odbacimo nul-hipotezu.

---

## 1. Test normalnosti reziduala (Anderson–Darling test)

### Nul-hipoteza  
Reziduali su normalno distribuirani.

### Tumačenje p-vrednosti  
P-vrednost ovde predstavlja verovatnoću da bismo videli ovako odstupanje od normalne raspodele **ako su reziduali u stvarnosti zaista normalni**.

- Ako je p-vrednost **veća ili jednaka 0.05**, posmatrano odstupanje nije naročito neobično za normalnu raspodelu → zaključujemo da reziduali mogu biti normalni.
- Ako je p-vrednost **manja od 0.05**, posmatrana odstupanja su isuviše neuobičajena za normalnu raspodelu → zaključujemo da reziduali verovatno nisu normalni.

Ovo je ista logika kao kod novčića:  
Ako posmatrani rezultat spada među one koji bi se veoma retko desili kada je novčić pošten, onda posmatrani rezultat daje razlog da posumnjamo da je novčić pošten.
Ovde: **Ako rezultat koji smo dobili spada među one koje bi normalna distribucija vrlo retko generisala, onda taj rezultat daje razlog da posumnjamo u pretpostavku da su reziduali normalno distribuirani..**

---

## 2. Test jednakosti varijanse (Goldfeld–Quandt test)

### Nul-hipoteza  
Reziduali imaju jednaku varijansu (homoskedastičnost).

### Tumačenje p-vrednosti  
P-vrednost meri koliko je verovatno da bismo dobili razliku u varijansama bar toliko izraženu kao u podacima **ako reziduali u stvarnosti zaista imaju jednaku varijansu**.

- Ako je p-vrednost **veća ili jednaka 0.05**, razlike u varijansi nisu dovoljno velike da se smatraju neuobičajenim → prihvatamo da varijansa može biti jednaka.
- Ako je p-vrednost **manja od 0.05**, razlika u varijansama je prevelika da bi se očekivala pod pretpostavkom jednake varijanse → zaključujemo da varijansa verovatno nije jednaka.

I ovde važi ista ideja kao i kod novčića:  
**Ako posmatrana razlika u varijansi spada među one koje bi se veoma retko javile kada bi varijansa zaista bila jednaka, onda to predstavlja razlog da sumnjamo da je pretpostavka o jednakoj varijansi tačna.**

---

## Suština oba testa

U oba slučaja p-vrednost predstavlja meru „normalnosti“ posmatranih odstupanja u odnosu na neku početnu pretpostavku (nul-hipotezu).  
Kao i u primeru bacanja novčića:

- Velika p-vrednost znači da rezultat nije neobičan → nul-hipoteza se zadržava.
- Mala p-vrednost znači da je rezultat previše ekstreman → nul-hipoteza se odbacuje.

Na taj način p-vrednost nam daje kvantitativnu osnovu za donošenje odluke o tome da li pretpostavke regresionog modela stoje ili ne.
